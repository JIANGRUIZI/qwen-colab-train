```python
keywords = [
    "嵌入式", "芯片", "MCU", "控制", "设备", "系统", "单片机",
    "微控制器", "传感器", "驱动", "RTOS", "总线", "硬件"
]

def is_valid(sample):
    instruction = sample["instruction"]
    output = sample["output"]
    if not instruction or not output:
        return False
    if not any(kw in instruction for kw in keywords):
        return False
    if not (30 <= len(output) <= 512):
        return False
    return True

filtered_dataset = dataset.filter(is_valid)
print(f"筛选后数据量: {len(filtered_dataset)} 条")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from datasets import load_dataset
import pandas as pd

# ✅ 基础配置
model_id = "Qwen/Qwen-1_8B-Chat"
token =  # 替换为你的token

# ✅ 加载 tokenizer 和模型
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True, token=token)
tokenizer.pad_token = "<|endoftext|>"  # Qwen 特有 end token 作为 pad
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True  # ✅ 新加这行！
)


model = AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    quantization_config=bnb_config,
    device_map={"": 0},
    token=token
)
model.config.pad_token_id = tokenizer.convert_tokens_to_ids("<|endoftext|>")
model.config.use_cache = False  # 🚨 关闭缓存，加快训练速度
model = prepare_model_for_kbit_training(model)

# ✅ 挂载 LoRA
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["c_attn", "W_pack", "q_proj", "v_proj", "k_proj", "o_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, lora_config)

# ✅ 加载格式化数据（你已保存好的 json）
dataset = load_dataset("json", data_files="/content/Qwen/filtered_belle_embedded.json", split="train")

# ✅ 格式化 prompt+response
def format_example(example):
    prompt = example["instruction"]
    if example.get("input"):
        prompt += "\n" + example["input"]
    return {"prompt": prompt, "response": example["output"]}

formatted_dataset = dataset.map(format_example)

# ✅ Tokenize + 添加 labels
def tokenize(example):
    tokenized = tokenizer(
        example["prompt"],
        text_target=example["response"],
        max_length=512,
        truncation=True,
        padding="max_length"
    )
    tokenized["labels"] = tokenized["input_ids"].copy()  # 👈 非常关键！
    return tokenized

tokenized_dataset = formatted_dataset.map(tokenize, remove_columns=["instruction", "input", "output", "prompt", "response"])

# ✅ 构造 collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

# ✅ 设置训练参数
training_args = TrainingArguments(
    output_dir="/content/qwen-lora-train",
    per_device_train_batch_size=24,
    gradient_accumulation_steps=4,
    learning_rate=2e-5,
    num_train_epochs=3,
    logging_steps=1,
    save_steps=10,
    save_total_limit=2,
    fp16=True,
    logging_dir="/content/logs",
    report_to="tensorboard"
)

# ✅ 初始化 Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer  # 虽然有 warning，但没关系
)

# ✅ 启动训练
trainer.train()
trainer.save_model("/content/qwen-lora-train/final-lora")

history = trainer.state.log_history
df = pd.DataFrame(history)
df.to_csv("/content/qwen-lora-train/loss_log.csv", index=False)
```

```python
import json
import torch
import random
import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# ✅ 模型路径配置
model_id = "Qwen/Qwen-1_8B-Chat"
adapter_path = "/content/final-lora"

# ✅ 加载 tokenizer 和模型
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
smooth_fn = SmoothingFunction().method1
rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

# ✅ 加载原始模型和微调模型
base_model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", trust_remote_code=True).eval()
lora_model = PeftModel.from_pretrained(base_model, adapter_path).eval()

# ✅ 替换为人工设计的两条样本
examples = [
    {
        "instruction": "简述嵌入式系统中 Flash 和 RAM 的作用区别。",
        "reference": (
            "Flash 是一种非易失性存储器，主要用于存放程序代码、启动固件和常量数据，即使在断电的情况下也能保留信息，"
            "常见于 MCU 的代码区；而 RAM 属于易失性存储器，主要用于运行时的数据缓存、变量存储和堆栈操作，其速度远高于 Flash，"
            "但在掉电后数据会全部丢失。两者在嵌入式系统中扮演不同角色：Flash 保证了系统的启动与逻辑执行基础，而 RAM 提供了必要的运行环境与高速数据访问能力。"
        )
    },
    {
        "instruction": "RTOS 中为什么使用抢占式调度？",
        "reference": (
            "在 RTOS 中使用抢占式调度的主要目的是保障系统实时性。抢占式调度允许高优先级任务在任意时间中断当前正在执行的低优先级任务，"
            "以确保关键任务在系统中可以被及时响应。这种机制特别适用于多任务并行处理的场景，如中断服务、信号采集和实时控制等。"
            "与非抢占式调度相比，抢占式调度具有更强的确定性和响应控制能力，能够降低任务延迟，减少关键任务等待时间，是嵌入式实时应用中广泛采用的调度策略。"
        )
    },
    {
        "instruction": "中断优先级在 MCU 中有何意义？",
        "reference": (
            "中断优先级用于定义当多个中断同时或几乎同时发生时，处理器应优先响应哪个中断请求。它通过设定不同中断源的优先级等级，确保关键性更高的事件能被优先处理，"
            "从而有效避免延迟和响应阻塞。例如，在一个嵌入式控制系统中，来自温度传感器的过热警告中断应优先于常规的串口通信中断。"
            "合理配置中断优先级有助于保障系统的实时性和稳定性，避免资源冲突和处理死锁，是设计高可靠性 MCU 系统的基础手段之一。"
        )
    }
]


# ✅ 回复生成函数
def generate_reply(model, tokenizer, instruction):
    inputs = tokenizer(instruction, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=512,
            no_repeat_ngram_size=3,
            do_sample=True,
            temperature=0.7
        )
    generated = output[0][inputs["input_ids"].shape[-1]:]
    return tokenizer.decode(generated, skip_special_tokens=True)

# ✅ 文本清洗函数
def clean_text(text):
    return text.replace("\n", "").replace("\r", "").strip()

# ✅ 开始测试
results = []
for ex in examples:
    instr = ex["instruction"].replace("\n", " ")
    ref = ex["reference"]
    prompt = instr

    base_out = generate_reply(base_model, tokenizer, prompt)
    lora_out = generate_reply(lora_model, tokenizer, prompt)

    base_out = clean_text(base_out)
    lora_out = clean_text(lora_out)

    if not base_out.strip():
        base_out = "[无回答]"
    if not lora_out.strip():
        lora_out = "[无回答]"

    bleu_base = sentence_bleu([ref], base_out, smoothing_function=smooth_fn)
    bleu_lora = sentence_bleu([ref], lora_out, smoothing_function=smooth_fn)
    rouge_base = rouge.score(ref, base_out)["rougeL"].fmeasure
    rouge_lora = rouge.score(ref, lora_out)["rougeL"].fmeasure

    results.append({
        "Instruction": instr,
        "Ref": ref,
        "Base Output": base_out,
        "LoRA Output": lora_out,
        "BLEU Base": round(bleu_base, 4),
        "BLEU LoRA": round(bleu_lora, 4),
        "ROUGE Base": round(rouge_base, 4),
        "ROUGE LoRA": round(rouge_lora, 4),
    })
# ✅ 清理显存
del base_model
del lora_model
torch.cuda.empty_cache()
torch.cuda.ipc_collect()

# ✅ 显示结果
df = pd.DataFrame(results)
df



```

```python
import json, random
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import pandas as pd
import torch

# ✅ 基础配置
model_id = "Qwen/Qwen-1_8B-Chat"
adapter_path = "/content/final-lora"

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
smooth_fn = SmoothingFunction().method1
rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

# ✅ 加载模型
base_model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", trust_remote_code=True).eval()
lora_model = PeftModel.from_pretrained(base_model, adapter_path).eval()

# ✅ 读取 JSONL 并抽样
data = []
with open("/content/Qwen/filtered_belle_embedded.json", "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line))
samples = random.sample(data, 3)

# ✅ 构建输入
examples = [{"instruction": s["instruction"], "reference": s["output"]} for s in samples]

# ✅ 生成函数
def generate_reply(model, instruction, max_tokens=512):
    inputs = tokenizer(instruction, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=max_tokens)
    return tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

# ✅ 比较并保存结果
results = []
for ex in examples:
    instr = ex["instruction"]
    ref = ex["reference"]

    base_out = generate_reply(base_model, instr)
    lora_out = generate_reply(lora_model, instr)

    bleu_base = sentence_bleu([ref], base_out, smoothing_function=smooth_fn)
    bleu_lora = sentence_bleu([ref], lora_out, smoothing_function=smooth_fn)
    rouge_base = rouge.score(ref, base_out)["rougeL"].fmeasure
    rouge_lora = rouge.score(ref, lora_out)["rougeL"].fmeasure

    results.append({
        "Instruction": instr,
        "Ref": ref,
        "Base Output": base_out,
        "LoRA Output": lora_out,
        "BLEU Base": round(bleu_base, 4),
        "BLEU LoRA": round(bleu_lora, 4),
        "ROUGE Base": round(rouge_base, 4),
        "ROUGE LoRA": round(rouge_lora, 4),
    })

del base_model
del lora_model
torch.cuda.empty_cache()
torch.cuda.ipc_collect()

df = pd.DataFrame(results)
df
```
